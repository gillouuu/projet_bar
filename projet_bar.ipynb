```python
s = Service('C:/Users/Corentin/Desktop/IA school/WEBscrapping/chromedriver-win64/chromedriver-win64/chromedriver.exe')
driver = webdriver.Chrome(service=s)


driver.get('https://www.schlouk-map.com/fr/places/tom-pouce')
button = driver.find_element(By.CLASS_NAME, "fc-button-label")
time.sleep(1)

button.click()

time.sleep(1)
page_source = driver.page_source
time.sleep(1)


soup = BeautifulSoup(page_source, 'html.parser')

ouverture_div = soup.find('div', {'id': 'ouverture'})
if ouverture_div:
    table = ouverture_div.find('table', class_='table table-sm table-borderless mb-0')


table_texts = []

# Parcourir les lignes de la table
rows = table.find_all('tr')
for row in rows:
    # Extraire les données de chaque cellule dans la ligne
    cells = row.find_all('td')
    for cell in cells:
        # Ajouter le texte de chaque cellule à la liste
        table_texts.append(cell.text.strip())

# Afficher la liste contenant tous les textes de la table
print(table_texts)
```

```python
horaires = []

for i in range(0, len(table_texts), 2):
    horaires.append(table_texts[i] + " : " + table_texts[i+1])

print(horaires)
```

```python
s = Service('C:/Users/Corentin/Desktop/IA school/WEBscrapping/chromedriver-win64/chromedriver-win64/chromedriver.exe')
driver = webdriver.Chrome(service=s)


driver.get('https://www.schlouk-map.com/fr/places/tom-pouce')
button = driver.find_element(By.CLASS_NAME, "fc-button-label")
time.sleep(1)

button.click()

time.sleep(1)
page_source = driver.page_source
time.sleep(1)


soup = BeautifulSoup(page_source, 'html.parser')


table_prices = soup.find('table', class_='table table-sm table-borderless table-drinks')



prices_text = []

rows = table_prices.find_all('tr')
for row in rows:
    cells = row.find_all('td')
    for cell in cells:
        prices_text.append(cell.text.strip())



prices_dict = {}

for i in range(0, len(prices_text), 3):
    nom = prices_text[i]
    happyhour = prices_text[i+1]
    tarif_normal = prices_text[i+2]
    
    price_info = {
        "nom": nom,
        "happyhour": happyhour,
        "tarif normal": tarif_normal
    }
    
    key = nom.lower() 
    prices_dict[key] = price_info

print(prices_dict)
```
